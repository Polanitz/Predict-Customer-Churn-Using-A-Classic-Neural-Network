# Creating Dummy Variables for Categorical Variables

```python
cat_vars=['Geography', 'Gender', 'NumOfProducts', 'HasCrCard', 'IsActiveMember']
for var in cat_vars:
    cat_list='var'+'_'+var
    cat_list = pd.get_dummies(df[var], prefix=var)
    df1=df.join(cat_list)
    df=df1
```

```python
df.drop(['Geography', 'Gender', 'NumOfProducts', 'HasCrCard', 'IsActiveMember'], axis=1, inplace=True)
df.columns.values
```

```python
df_vars=df.columns.values.tolist()
y = ['Churned']
X = [i for i in df_vars if i not in y]
```

```python
cols = X
cols
```

# Over-sampling using SMOTE

```python
X = df[cols]
y = df['Churned']

from imblearn.over_sampling import SMOTE

os = SMOTE(random_state=42)
columns = X.columns

os_data_X,os_data_y=os.fit_sample(X, y)
os_data_X = pd.DataFrame(data=os_data_X,columns=columns)
os_data_y = pd.DataFrame(data=os_data_y,columns=['Churned'])

# we can Check the numbers of our data
print("\033[1m Length of oversampled data is:",len(os_data_X))
print("\033[1m Number of churned in oversampled data is:",len(os_data_y[os_data_y['Churned']==1]))
print("\033[1m Number of no-churned in oversampled data is:",len(os_data_y[os_data_y['Churned']==0]))
print("\033[1m Proportion of churned data in oversampled data is:",len(os_data_y[os_data_y['Churned']==1])/len(os_data_X))
print("\033[1m Proportion of no-churned data in oversampled data is:",len(os_data_y[os_data_y['Churned']==0])/len(os_data_X))
```

# Data Processing For Neural Network Training

```python
from sklearn.model_selection import train_test_split
X_train1, X_test, y_train1, y_test = train_test_split(os_data_X, os_data_y, test_size=0.33, random_state=42, stratify=os_data_y)
```

```python
import numpy as np
from sklearn.utils import shuffle

train_labels = np.array(y_train1)
train_samples = np.array(X_train1)
train_labels, train_samples = shuffle(train_labels, train_samples)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform(train_samples)
```

```python
for i in scaled_train_samples:
    print(i)
```

# Creating An Artificial Neural Network

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

```python
X_train1 = scaled_train_samples
y_train1 = train_labels
```

```python
model = Sequential([
    Dense(units=16, input_shape=(X_train1.shape[1],), activation='relu'),
    Dense(units=32, activation='relu'),
    Dense(units=2, activation='softmax')
])
```

```python
model.summary()
```

# Building A Validation Set

```python
X_train, x_val, y_train, y_val = train_test_split(X_train1, y_train1, test_size=0.10, random_state=42)
```

```python
X_train.shape
```

```python
x_val.shape
```

```python
valid_set = (x_val, y_val)
```

# Training An Artificial Neural Network

```python
model.compile(optimizer=Adam(learning_rate=0.0001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x=X_train, y=y_train, validation_data=valid_set, batch_size=10, epochs=30, verbose=2)
```

# Predictions Using An Artificial Neural Network

```python
test_labels = np.array(y_test)
test_samples = np.array(X_test)
test_labels, test_samples = shuffle(test_labels, test_samples)
scaled_test_samples = scaler.fit_transform(test_samples)
```

```python
X_test = scaled_test_samples
y_test = test_labels
```

```python
predictions = model.predict(x=X_test, batch_size=10, verbose=0)  
```

```python
for i in predictions:
    print(i)
```

```python
rounded_predictions = np.argmax(predictions, axis=-1)

for i in rounded_predictions:
    print(i)
```

# Creating A Confusion Matrix For the Predictions

```python
%matplotlib inline
from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt

y_true = test_labels
y_pred = rounded_predictions
cm = confusion_matrix(y_true=y_true, y_pred=y_pred)
```

```python
def plot_confusion_matrix(cm, classes,
                        normalize=False,
                        title='Confusion matrix',
                        cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
            horizontalalignment="center",
            color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
cm_plot_labels = ['No-Churned', 'Churned']
```

```python
plot_confusion_matrix(cm=cm, classes=cm_plot_labels, title='Confusion Matrix')
```

```python
print("\033[1mThe result is telling us that we have: ",(cm[0,0]+cm[1,1]),"correct predictions.")
print("\033[1mThe result is telling us that we have: ",(cm[0,1]+cm[1,0]),"incorrect predictions.")
print("\033[1mWe have a total predictions of: ",(cm.sum()))
```

# Building The ROC Curve

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
import seaborn as sns
sns.set(style="darkgrid", color_codes=True)

ann_roc_auc = roc_auc_score(y_test,  model.predict(X_test)[:,1])
fpr, tpr, thresholds = roc_curve(y_test, model.predict_proba(X_test)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Classic Neural Network (area = %0.4f)' % ann_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('CNN_ROC')
plt.show()
```
